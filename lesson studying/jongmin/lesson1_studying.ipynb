모듈 부르고 시스템 형성

```python
import pychrono as chrono
sys = chrono.ChSystemNSC()                                       #물리세계 생성
sys.SetGravitationalAcceleration(chrono.ChVector3d(0, -9.81, 0)) #중력가속도
print("물리 시스템 생성 완료!")
print(f"  중력: {sys.GetGravitationalAcceleration().y} m/s²")
```

물체 생성

```python
ball = chrono.ChBody() # 강체 생성
ball.SetMass(1.0)                           # 1 kg 질량 
ball.SetPos(chrono.ChVector3d(0, 10, 0))    # 초기 높이 10m
ball.SetFixed(False)                        # 고정하지 않음 → 자유롭게 움직임
sys.AddBody(ball)                           # 세계에 공 넣기

print(f"\n공 생성 완료!")
print(f"  질량: {ball.GetMass()} kg")
print(f"  초기 위치: y = {ball.GetPos().y} m")
print(f"  고정 여부: False (자유 낙하)")
```

시뮬 시작

```python
dt = 0.01   # 시간 간격을 0.01초로

print(f"\n{'시간(s)':>8}  {'높이(m)':>10}  {'속도(m/s)':>10}  {'이론높이(m)':>12}")
print("─" * 48)

while sys.GetChTime() < 2.0:     # 2초동안 자유 낙하 보기
    sys.DoStepDynamics(dt)       # 시뮬을 dt만큼 진행, 반드시 필요함

    t = sys.GetChTime()          
    y = ball.GetPos().y          # 현재 높이
    vy = ball.GetPosDt().y       # 현재 y방향 속도 (GetPosDt = 위치의 시간 미분 = 속도)

    # 이론값: y = y0 - (1/2) * g * t²
    y_theory = 10.0 - 0.5 * 9.81 * t * t

    # 0.2초마다 출력
    if abs(t % 0.2) < dt * 0.5 or abs(t % 0.2 - 0.2) < dt * 0.5:
        print(f"{t:8.2f}  {y:10.4f}  {vy:10.4f}  {y_theory:12.4f}")
```

결과 확인

```python
final_y = ball.GetPos().y            # 2초 후 시뮬 높이
theory_y = 10 - 0.5 * 9.81 * 2.0**2  # 2초 후 이론 높이

print(f"\n{'='*48}")
print(f"시뮬레이션 결과: y = {final_y:.4f} m")
print(f"이론값 (물리):   y = {theory_y:.4f} m")
print(f"오차:            {abs(final_y - theory_y):.6f} m")  # 오차 계산
print(f"{'='*48}")
print()
print("축하합니다! 첫 번째 물리 시뮬레이션을 완료했습니다!")
print()
print("다음 단계:")
print("  → lesson_02에서 바닥과의 충돌을 추가해 봅시다")
print("  → lesson_03에서 3D 시각화를 추가해 봅시다")
```
