모듈 부르기

```python
import pychrono as chrono
import pychrono.irrlicht as chronoirr  # 3D 그래픽 엔진

print("=" * 55)
print("Lesson 03: 3D 시각화 (Irrlicht)")
print("=" * 55)
```

```python
sys = chrono.ChSystemNSC()
sys.SetGravitationalAcceleration(chrono.ChVector3d(0, -9.81, 0))
sys.SetCollisionSystemType(chrono.ChCollisionSystem.Type_BULLET)

# 접촉 재질
material = chrono.ChContactMaterialNSC()
material.SetFriction(0.4)                 # 마찰 계수 0.4
material.SetRestitution(0.5)              # 반발 계수 0.5
```

바닥 생성

```python
floor = chrono.ChBody()
floor.SetPos(chrono.ChVector3d(0, -1, 0))
floor.SetFixed(True)
floor.EnableCollision(True)   # 충돌 켜기
floor.AddCollisionShape(chrono.ChCollisionShapeBox(material, 20, 2, 20))  # 충돌 모양이 아니라 보이는 모양

floor_shape = chrono.ChVisualShapeBox(20, 2, 20)
floor_shape.SetColor(chrono.ChColor(0.4, 0.5, 0.4))  # 바닥색 녹색으로 설정
floor.AddVisualShape(floor_shape)                    # 바닥을 시각적 모양으로

sys.AddBody(floor)
print("바닥 생성 (녹색 상자)")
```

시뮬 시작

```python
def create_ball(system, mat, pos, radius, color, name):     # 공 3개 생성
    ball = chrono.ChBody()
    ball.SetMass(1.0)
    ball.SetPos(pos)     # 공마다 다른 높이롷 시작, 함수 인자
    ball.SetFixed(False)
    ball.EnableCollision(True)

    ball.AddCollisionShape(chrono.ChCollisionShapeSphere(mat, radius)) # 충돌 모양 구

    inertia = 2.0 / 5.0 * ball.GetMass() * radius**2
    ball.SetInertiaXX(chrono.ChVector3d(inertia, inertia, inertia))

    ball_shape = chrono.ChVisualShapeSphere(radius)   # 시각적 모양이 구
    ball_shape.SetColor(color)                        # 색 설정
    ball.AddVisualShape(ball_shape)

    system.AddBody(ball)
    print(f"  {name}: 위치=({pos.x}, {pos.y}, {pos.z})")
    return ball                              # 생성한 공 함수 밖으로 돌려주기

print("\n공 3개 생성:")

ball_red = create_ball(       # 각 공 설정
    sys, material,
    chrono.ChVector3d(-2, 8, 0), 0.4,
    chrono.ChColor(0.9, 0.2, 0.2), "빨간 공"
)

ball_blue = create_ball(      # 각 공 설정
    sys, material,
    chrono.ChVector3d(0, 5, 0), 0.3,
    chrono.ChColor(0.2, 0.3, 0.9), "파란 공"
)

ball_yellow = create_ball(    # 각 공 설정
    sys, material,
    chrono.ChVector3d(2, 3, 0), 0.35,
    chrono.ChColor(0.9, 0.8, 0.1), "노란 공"
)
```

시각화 시스템 생성

```python
print("\n3D 시각화 초기화 중...")                   # 시각화 시스템 생성
vis = chronoirr.ChVisualSystemIrrlicht()        
vis.AttachSystem(sys)                            # 물리 시스템과 화면 시스템 연결
vis.SetWindowSize(1280, 720)
vis.SetWindowTitle('Lesson 03: Bouncing Balls')  
vis.Initialize()                                 # 시각화 시스템 초기화

vis.AddSkyBox()                                  # 하늘 배경 추가

vis.AddCamera(                        # 카메라 추가
    chrono.ChVector3d(0, 5, 12),      # 카메라 위치
    chrono.ChVector3d(0, 2, 0)        # 바라보는 위치
)

vis.AddTypicalLights()                # 기본 조명

print("시각화 준비 완료!")
```

시뮬 루프

```python
step_size = 0.005     # dt랑 같음

while vis.Run():                   # 창 열린 기간 시뮬 반복
    vis.BeginScene()               # 한 프레임 그리기 시작
    vis.Render()                   # 렌더링
    vis.EndScene()                 # 그리기 종료
    sys.DoStepDynamics(step_size)  # 
```

결과 정리

```python
print(f"\n시뮬레이션 종료! (총 {sys.GetChTime():.2f}초 실행)")
print(f"  빨간 공 최종 높이: {ball_red.GetPos().y:.3f} m")
print(f"  파란 공 최종 높이: {ball_blue.GetPos().y:.3f} m")
print(f"  노란 공 최종 높이: {ball_yellow.GetPos().y:.3f} m")
```
